```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your Netflix dataset
file_path = "netflix_titles.csv"
df = pd.read_csv(file_path)

# Filter only movies (since duration in minutes applies to movies)
movies = df[df['type'] == 'Movie'].copy()

# Convert 'duration' column to numeric (remove 'min' text)
movies['duration_min'] = movies['duration'].str.replace(' min', '', regex=False).astype(float)

# Create histogram chart (Matplotlib version)
plt.figure(figsize=(10,6))
plt.hist(
    movies['duration_min'].dropna(),
    bins=20,                      # number of bars
    color='skyblue',
    edgecolor='black',
    alpha=0.7
)

# Add titles and labels
plt.title('Histogram: Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()
genre_counts = genres.value_counts()

# Take the top N genres for better visualization
top_genres = genre_counts.head(20)

# Create a bar plot of the top genres
plt.figure(figsize=(12, 6))
plt.bar(top_genres.index, top_genres.values)
plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Distribution of Content by Top 20 Genres")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
